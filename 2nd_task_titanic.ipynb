```python
#filling age column with mean 
```

```python
a = df['age'].mean()
df['age'] = df['age'].fillna(a)
```

```python
df
```

```python
# alive and adult_male column has categorical values 
# assigning 0 to no and 1 to yes
```

```python
df['alive'] = np.where(df['alive']=='no',0,1)
```

```python
df
```

```python
df['adult_male'] = np.where(df['adult_male']== False,0,1)
```

```python
sns.barplot(x = 'sex',y = 'alive',data = df)
```

```python
# survival rate by age 
```

```python
df['age_group'] = pd.cut(df['age'],
                        bins = [0,12,18,30,50,80],
                        labels = ['Child','Teen','Young','Adult','Senior']) 
```

```python
sns.barplot(x = 'age_group',y = 'alive',data = df )
```
